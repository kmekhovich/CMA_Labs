```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x = [500, 1000, 2000, 4000]

y0=[52.106, 193.028, 685.99, 1755.05]
y1=[6.173, 25.353, 97.224, 441.478]


fig = plt.figure(figsize=(10, 5))

plt.title("Gauss-Seidel & GMRES")
plt.plot(x, y0, label="Gauss-Seidel")
plt.plot(x, y1, label="GMRES")

plt.legend()

plt.show()

```
