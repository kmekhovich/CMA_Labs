```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x = [500, 1000, 2000, 4000]

y0=[16.161, 73.103, 309.817, 1106.78]
y1=[2.798, 13.891, 56.955, 427.814]


fig = plt.figure(figsize=(10, 5))

plt.title("Gauss-Seidel & GMRES")
plt.plot(x, y0, label="Gauss-Seidel")
plt.plot(x, y1, label="GMRES")

plt.legend()

plt.show()

```
